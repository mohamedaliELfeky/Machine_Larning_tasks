# Import 

```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as mtp
```

# Read Data of X and Y

```python
Data_frameX =  pd.read_csv('/home/el-feky/Downloads/X.csv')
Data_frameY = pd.read_csv('/home/el-feky/Downloads/Y.csv')
```

# Apply Equation Here and Store it at (theta) variable

```python
x1=np.array(Data_frameX)
y1=np.array(Data_frameY)

x_trans = x1.transpose()

transX = np.dot(x_trans, x1)
transY = np.dot(x_trans, y1)


thita = np.dot(np.linalg.inv(transX),transY)
thita
```

# Plot (OverallQual) on X axis and (SalePrice) On Y axis 

```python
mtp.plot(Data_frameX['OverallQual'],Data_frameY['SalePrice'])
mtp.title('House prices')
mtp.show()
```
